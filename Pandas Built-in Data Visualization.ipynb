We can use style sheets for an even nicer layout:

# Style Sheets
Matplotlib has [style sheets](https://matplotlib.org/gallery.html#style_sheets) to make your plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They make all your plots have the same look and feel more professional.

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("bmh")
df1['A'].hist(bins=30)
```

```python
plt.style.use("fivethirtyeight")
df1['A'].hist(bins=30)
```

```python
plt.style.use("dark_background")
df1['A'].hist(bins=30)
```

```python
plt.style.use("ggplot")
df2.plot(kind='scatter',x='a',y='b')
```

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

or just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___

# Area

```python
df2.plot.area(alpha=0.2)
plt.legend(bbox_to_anchor=(1, 0.75))
```

# Barplots

```python
sns.set(style='dark')
df2.plot.bar(cmap='magma')
```

```python
df2.plot.bar(x='c',y='b',cmap='coolwarm') #color='green'
```

```python
df2.plot.bar(stacked=True)
```

```python
df2.plot.bar(rot=45)
```

```python
df2.plot.barh(rot=45)
```

# Histograms

```python
df1['B'].plot.hist(bins=30)
```

```python
df1['B'].plot.hist(bins=25,stacked=False,alpha=0.5)
```

# Line plots

```python
df2.plot.line(x='a',y='b',figsize=(12,3),lw=2)
```

# Scatter Plots

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

You can use c to color based off another column value Use cmap to indicate colormap to use. For all the colormaps, check out:

```python
df1.plot.scatter(x='A',y='B',c='C',colormap='ocean',s=df1['C']*150,figsize=(12,6))
```

# BoxPlots

```python
df2.plot(kind='box',figsize=(10,5))
```

# Hexagonal Plot

Useful for Bivariate data

```python
df2.plot.hexbin(x='a',y='b',cmap='terrain')
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

# Pie chart

```python
df2.plot.pie(subplots=True,figsize=(20,10))
```

For a specific column:

```python
df2.plot.pie(y='c',cmap='coolwarm',figsize=(10,6))
```

# KDE

```python
df1['A'].plot.kde(lw=3)
```

```python
df1['A'].plot.density(lw=3,linestyle=':')
```

```python
df2.plot.kde(lw=3,figsize=(10,5))
```

This method of plotting will be a lot easier to use than full-on matplotlib, it balances ease of use with control over the figure. A lot of the plot calls also accept additional arguments of their parent matplotlib plt. call.
