Cufflinks connects plotly with pandas.

````python
from plotly.offline import download_plotlyjs , init_notebook_mode , plot , iplot
````

````python
import cufflinks as cf
````

The following command is going to connect the javascript to your notebook. Because plotly essentially just connects pandas and Python to its interactive javascript library.

And this is going to allow your notebooks to access those visualizations.

````python
init_notebook_mode(connected=True)
````

````python
# For cufflinks offline use
cf.go_offline()
````

````python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
````

````python
df.head()
````

````python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
````

````python
df2.head()
````

````python
df.plot() # Plot gives the usual, non-interactive plot
````

````python
df.iplot()   # Plots the interactive plot with value points,zoom,etc
````

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

# Scatter

````python
df.iplot(kind='scatter',x='A',y='B')
````

To convert this above line plot to actual scatter plot, we hve to use argument mode='markers'

````python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=15)
````

# Barplot

````python
df2.iplot(kind='bar',x='Category',y='Values')
````

If we work on some continuous data like there exists in df then we will get a bar plot like

````python
df.iplot(kind='bar')
````

What we can do is apply some aggreagate function on the data in order to get a single bar for a single column instead of a bar for each continuous value like one we are getting above

````python
df.count().iplot(kind='bar')
````

````python
df.sum().iplot(kind='bar')
````

# Boxplots

````python
df.iplot(kind='box')
````

# 3D surface data plot

Creating a dataframe with 3 dimensions

````python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[1,2,3,4,5]})
````

````python
df3.iplot(kind='surface',colorscale = 'rdpu')
````

````python
cf.colors.scales()
````

# Histograms

````python
df.iplot(kind='hist',bins=50)
````

````python
df['A'].iplot(kind='hist',bins=30,colorscale='pastel2')
````

# Spread plot

Useful for stock data as gives the difference in between constituent graphs values

````python
df[['A','B']].iplot(kind='spread')
````

# Bubble plot

Similar to a scatter plot with the difference that the size of the marker will change as per another variable

````python
df.iplot(kind='bubble',x='A',y='B',size='C')
````

# Scatter matrix

Similar to seaborn's pair plot . Works on numeric data

````python
df.scatter_matrix()
````
